```python
import numpy as np
import webbrowser
from aocd.models import Puzzle
import re
import itertools
from scipy.ndimage.filters import minimum_filter
from scipy.ndimage import label
import matplotlib.pyplot as plt
%matplotlib inline
puzzle = Puzzle(year=2021, day=9)
```

```python
webbrowser.open(puzzle.url);
```

## Part 1

```python
input_data = puzzle.input_data
```

```python
matrix = np.asarray([[int(num) for num in line] for line in input_data.split()])
minima = minimum_filter(matrix, size=3)
low_points = matrix[matrix == minima]
```

```python
plt.imshow(matrix, interpolation="none", cmap="RdYlBu_r")
plt.title("Heightmap")
plt.gcf().set_size_inches(12, 12)
```

```python
result = (low_points + 1).sum()
```

```python
puzzle.answer_a = result
```

## Part 2

```python
edges = (matrix == 9)
```

```python
labelled_basins, n = label(~edges)
```

```python
plt.imshow(labelled_basins, interpolation="none", cmap="RdYlBu_r")
plt.title("Basins")
plt.gcf().set_size_inches(12, 12);
```

```python
sizes = []
for i in range(1, n + 1):
    sizes.append((labelled_basins == i).sum())
```

```python
sizes.sort(reverse=True)
```

```python
result = np.prod(sizes[:3])
result
```

```python
puzzle.answer_b = result
```
